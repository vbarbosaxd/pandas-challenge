## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#First I have to extract the age column from the original dataframe
ad = purchase_data[['Age', 'SN']]

#Then I have to remove the duplicate SNs, so I only get one value per player
ad.drop_duplicates(subset ='SN',
                     keep = "first", inplace = True)
ad 
```

```python
#Establish my bins and bin names

bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
bin_age_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#Categorize the existing players using the age bins with pd.cut()
ad["Age Bins"] = pd.cut(ad["Age"], bins, labels=bin_age_names, include_lowest=True)
ad
```

```python
#Find total count by bin
ad_ser = ad["Age Bins"].value_counts()
ad_ser
```

```python
ad_df = pd.DataFrame(ad_ser)
ad_df
```

```python
ad_df = ad_df.rename(columns={'Age Bins': 'Total Count'})
ad_df
```

```python
#Percentage of players by age bin
adp = (ad_ser/p)*100
adp
```

```python
adp_df = pd.DataFrame(adp)
adp_df
```

```python
adp_df = adp_df.rename(columns={'Age Bins': 'Percentage of Players'})
adp_df
```

```python
Age_Demographics = pd.merge(ad_df, adp_df, left_index=True, right_index=True)
Age_Demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
ts = purchase_data[['SN', 'Price']]
ts
```

```python
ts_pc = ts.groupby('SN').count()
ts_pc = ts_pc.rename(columns={'Price': 'Purchase Count'})
ts_pc
```

```python
ts_app = ts.groupby('SN').mean()
ts_app = ts_app.rename(columns={'Price': 'Average Purchase Price'})
ts_app
```

```python
ts_tpv = ts.groupby('SN').sum()
ts_tpv = ts_tpv.rename(columns={'Price': 'Total Purchase Value'})
ts_tpv
```

```python
Top_Spenders = ts_pc.merge(ts_app, left_index=True, right_index=True).merge(ts_tpv, left_index=True, right_index=True)
Top_Spenders_df = Top_Spenders.sort_values(by='Total Purchase Value', ascending=False)
Top_Spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_pop = purchase_data[["Item ID", "Item Name", "Price"]]
most_pop
```

```python
#Find item purchase count
most_pop_pc = most_pop.groupby(["Item ID", "Item Name"]).count()
most_pop_pc = most_pop_pc.rename(columns={'Price': 'Purchase Count'})
most_pop_pc
```

```python
#Find Item Price
most_pop_ip = most_pop.groupby(["Item ID", "Item Name"]).mean()
most_pop_ip
```

```python
#Find Total Purchase Value
tpv_most_pop = most_pop.groupby(["Item ID", "Item Name"]).sum()
tpv_most_pop = tpv_most_pop.rename(columns={'Price': 'Total Purchase Value'})
tpv_most_pop
```

```python
Most_Pop_Items = most_pop_pc.merge(most_pop_ip, left_index=True, right_index=True).merge(tpv_most_pop, left_index=True, right_index=True)
Most_Pop_Items_df = Most_Pop_Items.sort_values(by='Purchase Count', ascending=False)
Most_Pop_Items_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
Most_Profitable_Items_df = Most_Pop_Items_df.sort_values(by='Total Purchase Value', ascending=False)
Most_Profitable_Items_df
```
